```python
#import bibliotecas
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import math
```

```python
#Média

# Lista de números
numeros = [3, 5, 7, 10, 15]

# Calculando a soma e o número de elementos
soma = sum(numeros)
quantidade = len(numeros)

# Calculando a média
media_mao = soma / quantidade
print(f"media na mão: {media_mao}")

# Calculando a média com NumPy
media = np.mean(numeros)

print(f"media com numpy: {media}")
```

```python
#Média Ponderada

# Lista de números e seus respectivos pesos
numeros = [3, 5, 7, 10, 15]
pesos = [1, 2, 3, 4, 5]  # Pesos para cada número

# Média ponderada na mão
numerador = sum(x * p for x, p in zip(numeros, pesos))
denominador = sum(pesos)
media_ponderada_mao = numerador / denominador

print(f"Média ponderada na mão: {media_ponderada_mao}")

# Média ponderada com NumPy
media_ponderada_numpy = np.average(numeros, weights=pesos)

print(f"Média ponderada com NumPy: {media_ponderada_numpy}")
```

```python
#Moda

# Lista de números
numeros = [3, 5, 7, 10, 15, 7, 7, 10]

# Calculando a moda na mão
frequencia = {}
for num in numeros:
    if num in frequencia:
        frequencia[num] += 1
    else:
        frequencia[num] = 1

# Encontrando o valor com a maior frequência
moda_mao = max(frequencia, key=frequencia.get)

print(f"Moda na mão: {moda_mao}")

# Calculando a moda com SciPy
moda_scipy = stats.mode(numeros, keepdims=True)[0][0]

print(f"Moda com SciPy: {moda_scipy}")
```

```python
#Percentil

# Lista de números
numeros = [3, 5, 7, 10, 15, 7, 7, 10]

# Ordenando os números
numeros_ordenados = sorted(numeros)

# Definindo o percentil (por exemplo, o percentil 50, que é a mediana)
percentil_desejado = 50

# Calculando a posição correspondente ao percentil
index = (percentil_desejado / 100) * (len(numeros_ordenados) - 1)
index_inferior = int(index)
index_superior = index_inferior + 1

# Se a posição é um número inteiro, usamos diretamente o valor; caso contrário, interpolamos entre os dois valores
if index_inferior == index:
    percentil_mao = numeros_ordenados[index_inferior]
else:
    peso = index - index_inferior
    percentil_mao = numeros_ordenados[index_inferior] * (1 - peso) + numeros_ordenados[index_superior] * peso

print(f"Percentil {percentil_desejado} na mão: {percentil_mao}")

# Calculando o percentil com NumPy
percentil_numpy = np.percentile(numeros, percentil_desejado)

print(f"Percentil {percentil_desejado} com NumPy: {percentil_numpy}")
```

```python
#Quartil

# Lista de números
numeros = [3, 5, 7, 10, 15, 7, 7, 10]

# Ordenando os números
numeros_ordenados = sorted(numeros)

def calcular_percentil_mao(percentil_desejado, numeros_ordenados):
    # Calculando a posição correspondente ao percentil
    index = (percentil_desejado / 100) * (len(numeros_ordenados) - 1)
    index_inferior = int(index)
    index_superior = index_inferior + 1

    # Se a posição é um número inteiro, usamos diretamente o valor; caso contrário, interpolamos entre os dois valores
    if index_inferior == index:
        return numeros_ordenados[index_inferior]
    else:
        peso = index - index_inferior
        return numeros_ordenados[index_inferior] * (1 - peso) + numeros_ordenados[index_superior] * peso

# Calculando os quartis
Q1_mao = calcular_percentil_mao(25, numeros_ordenados)
Q2_mao = calcular_percentil_mao(50, numeros_ordenados)  # Mediana
Q3_mao = calcular_percentil_mao(75, numeros_ordenados)

print(f"Quartil 1 na mão (Q1): {Q1_mao}")
print(f"Quartil 2 na mão (Q2): {Q2_mao}")
print(f"Quartil 3 na mão (Q3): {Q3_mao}")


# Calculando os quartis com NumPy
Q1_numpy = np.percentile(numeros, 25)
Q2_numpy = np.percentile(numeros, 50)  # Mediana
Q3_numpy = np.percentile(numeros, 75)

print(f"Quartil 1 com NumPy (Q1): {Q1_numpy}")
print(f"Quartil 2 com NumPy (Q2): {Q2_numpy}")
print(f"Quartil 3 com NumPy (Q3): {Q3_numpy}")
```

```python
#Amplitude

# Lista de números
numeros = [3, 5, 7, 10, 15, 7, 7, 10]

# Calculando a amplitude na mão
valor_maximo = max(numeros)
valor_minimo = min(numeros)
amplitude_mao = valor_maximo - valor_minimo

print(f"Amplitude na mão: {amplitude_mao}")

# Calculando a amplitude com NumPy
amplitude_numpy = np.max(numeros) - np.min(numeros)

print(f"Amplitude com NumPy: {amplitude_numpy}")
```

```python
#Variancia

# Lista de números
numeros = [3, 5, 7, 10, 15, 7, 7, 10]

# Calculando a média
media = sum(numeros) / len(numeros)

# Calculando a variância na mão
soma_diferencas_quadrado = sum((x - media) ** 2 for x in numeros)
variancia_mao = soma_diferencas_quadrado / len(numeros)

print(f"Variância na mão: {variancia_mao}")

# Calculando a variância com NumPy
variancia_numpy = np.var(numeros)

print(f"Variância com NumPy: {variancia_numpy}")
```

```python
#Desvio Padrão

numeros = [3, 5, 7, 10, 15, 7, 7, 10]

# Calculando a média
media = sum(numeros) / len(numeros)

# Calculando a soma das diferenças ao quadrado
soma_diferencas_quadrado = sum((x - media) ** 2 for x in numeros)

# Calculando a variância
variancia_mao = soma_diferencas_quadrado / len(numeros)

# Calculando o desvio padrão na mão
desvio_padrao_mao = variancia_mao ** 0.5

print(f"Desvio Padrão na mão: {desvio_padrao_mao}")

# Calculando o desvio padrão com NumPy
desvio_padrao_numpy = np.std(numeros)

print(f"Desvio Padrão com NumPy: {desvio_padrao_numpy}")
```

```python
#Distribuição normal 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros da distribuição normal
media = 0
desvio_padrao = 1

# Gerando valores de x (entre -3 e 3)
valores_x = np.linspace(-3, 3, 1000)

# Usando a função predefinida de densidade de probabilidade normal (PDF)
densidade_normal = norm.pdf(valores_x, media, desvio_padrao)

# Plotando a distribuição normal
plt.plot(valores_x, densidade_normal, label='Distribuição Normal')
plt.title('Distribuição Normal')
plt.xlabel('Valor')
plt.ylabel('Densidade de Probabilidade')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Erro Padrão

# Lista de números
numeros = [3, 5, 7, 10, 15, 7, 7, 10]

# Calculando a média
media = sum(numeros) / len(numeros)

# Calculando a soma das diferenças ao quadrado
soma_diferencas_quadrado = sum((x - media) ** 2 for x in numeros)

# Calculando a variância amostral (dividido por n-1)
variancia_amostral = soma_diferencas_quadrado / (len(numeros) - 1)

# Calculando o desvio padrão amostral
desvio_padrao_amostral = math.sqrt(variancia_amostral)

# Calculando o erro padrão
n = len(numeros)
erro_padrao_mao_amostral = desvio_padrao_amostral / math.sqrt(n)

print(f"Erro Padrão na mão (amostral): {erro_padrao_mao_amostral}")

# Calculando o erro padrão com SciPy
erro_padrao_scipy = stats.sem(numeros)

print(f"Erro Padrão com SciPy: {erro_padrao_scipy}")
```
